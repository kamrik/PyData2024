### Load IRIS dataset

```python
# pip install scikit-learn
from sklearn import datasets
iris = datasets.load_iris()
```

```python
print(iris.feature_names)
print(iris.target_names)
```

```python
print(iris.data)
print(iris.target)
```

```python
import numpy as np
X = np.array(iris.data)
y = np.array(iris.target)
print(X.shape)
print(y.shape)
```

### Exploratory Data Analysis

Box plot for each feature

```python
#box plot for each feature of the iris dataset
import matplotlib.pyplot as plt
plt.figure()
plt.boxplot(X)
plt.xticks(range(1,5), iris.feature_names)
plt.show()
```

Scatter plot for each pair of features

```python
import matplotlib.pyplot as plt

def my_plot(x1, x2, x1_label , x2_label, target_names):
    fig, ax = plt.subplots()
    scatter = ax.scatter(x1, x2 , c=target_names)
    ax.set(xlabel=x1_label, ylabel=x2_label)
    fig = ax.legend(
        scatter.legend_elements()[0], iris.target_names, loc="lower right", title="Classes"
        )

my_plot(X[:,0], X[:,1], iris.feature_names[0], iris.feature_names[1], y)
my_plot(X[:,0], X[:,2], iris.feature_names[0], iris.feature_names[2], y)
my_plot(X[:,0], X[:,3], iris.feature_names[0], iris.feature_names[3], y)
my_plot(X[:,1], X[:,2], iris.feature_names[1], iris.feature_names[2], y)
my_plot(X[:,1], X[:,3], iris.feature_names[1], iris.feature_names[3], y)
my_plot(X[:,2], X[:,3], iris.feature_names[2], iris.feature_names[3], y)
```

Histogram of each feature

```python
# histogram for each feature of the iris dataset
import matplotlib.pyplot as plt
plt.figure()
plt.hist(X[:,0], bins=20, edgecolor='black')
plt.xlabel(iris.feature_names[0])
plt.show()

plt.figure()
plt.hist(X[:,1], bins=20, edgecolor='black')
plt.xlabel(iris.feature_names[1])
plt.show()

plt.figure()
plt.hist(X[:,2], bins=20, edgecolor='black')
plt.xlabel(iris.feature_names[2])
plt.show()
```
