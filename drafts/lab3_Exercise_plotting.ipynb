```python
# Here we use SKLearn to load the iris dataset only
# see the cell below if you don't have SKLearn installed
from sklearn import datasets
iris = datasets.load_iris()
```

### Exploratory Data Analysis

The interesting parts of the `iris` object:
 - `iris.target_names` - the names of the 3 species of iris flowers
 - `iris.feature_names` - names of the different measurements performed on each flower
 - `iris.target` - the species of each iris flower in the dataset, it's given as a number 0, 1 or 2. iris.target_names[0] is the name of species 0.
 - `iris.data` - the actual measurements, in this case the length and the width in centimeters of some parts of the flower, see the photo for details


The names of variables may sound weird, the logic here is that we will be trying to predict the `target` knowing only the measurements in the `data`. This naming is common in ML.

```python
# The 3 species of Iris flowers in this dataset
for i, name in enumerate(iris.target_names):
    print(f'Species {i} - {name}')
```

```python
# The 4 measurements of the flowers
# See photo above or at this link https://s3.amazonaws.com/assets.datacamp.com/blog_assets/Machine+Learning+R/iris-machinelearning.png
for i, fname in enumerate(iris.feature_names):
    print(f'Feature {i} - {fname}')
```

```python
# It is common to name the data as X (upper case)
# and the thing we will be trying to figure out - the target as y (lower case)
X = iris.data
y = iris.target
```
